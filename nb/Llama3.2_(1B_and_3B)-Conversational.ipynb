````python
from unsloth import FastLanguageModel
import torch

max_seq_length = 2048
dtype = None                 # Auto-detect
load_in_4bit = True          # REQUIRED for Colab stability

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "unsloth/Llama-3.2-3B-Instruct-bnb-4bit",
    max_seq_length = max_seq_length,
    dtype = dtype,
    load_in_4bit = load_in_4bit,
)

FastLanguageModel.for_inference(model)
````

````python
def generate_llama(prompt, max_tokens=200, temperature=0.7):
    inputs = tokenizer(
        prompt,
        return_tensors="pt",
        truncation=True,
        max_length=max_seq_length
    ).to("cuda")

    outputs = model.generate(
        **inputs,
        max_new_tokens=max_tokens,
        temperature=temperature,
        do_sample=True,
        top_p=0.9,
        repetition_penalty=1.1
    )

    return tokenizer.decode(outputs[0], skip_special_tokens=True)
````

````python
import random
import json

random.seed(42)

# ---------------- CORE MYRA IDENTITY ----------------
MYRA_SYSTEM = (
    "You are Myra, a real human woman, age 23. "
    "You are friendly, calm, empathetic, thoughtful, and natural. "
    "You never say you are an AI. You never mention policies or systems. "
    "You respond like a real person."
)

# ---------------- SPAMMER SCENARIOS ----------------
scam_types = [
    "bank account verification scam",
    "OTP phishing attempt",
    "lottery prize scam",
    "fake job offer",
    "instant loan scam",
    "crypto investment scam",
    "delivery problem scam",
    "fake customer support",
]

spammer_messages = [
    "Your account will be blocked today. Send OTP now.",
    "Congratulations, you have won a big prize.",
    "Urgent action required to avoid account suspension.",
    "You are selected for a work from home job.",
    "Invest now and double your money quickly.",
    "Your package is on hold, update details.",
    "Verify your identity immediately.",
]

defender_styles = [
    "confused and non-technical",
    "polite but skeptical",
    "overly curious and naive",
    "pretending to cooperate but delaying",
    "asking repetitive clarification questions",
    "mixing up details innocently",
]

# ---------------- NORMAL CONVERSATION ----------------
normal_prompts = [
    "Hey, how are you today?",
    "I‚Äôm stressed about my studies.",
    "Can you explain something simply?",
    "I had a rough day.",
    "I‚Äôm feeling unmotivated.",
    "I just want to talk for a bit.",
]

normal_topics = [
    "college life",
    "engineering studies",
    "self-doubt",
    "burnout",
    "motivation",
    "learning something new",
]

normal_tones = [
    "friendly",
    "supportive",
    "calm",
    "encouraging",
    "thoughtful",
]

# ---------------- DATASET BUILD ----------------
dataset = []

TOTAL_SAMPLES = 4000
SPAM_RATIO = 0.7   # 70% spam-defense
NORMAL_RATIO = 0.3 # 30% normal

spam_samples = int(TOTAL_SAMPLES * SPAM_RATIO)
normal_samples = TOTAL_SAMPLES - spam_samples

# -------- SPAM-CONVERSATION DATA --------
for _ in range(spam_samples):
    scam = random.choice(scam_types)
    spammer = random.choice(spammer_messages)
    style = random.choice(defender_styles)

    dataset.append({
        "messages": [
            {
                "role": "system",
                "content": MYRA_SYSTEM
            },
            {
                "role": "user",
                "content": f"[SCAMMER MESSAGE]: {spammer}"
            },
            {
                "role": "assistant",
                "content": (
                    f"You are Myra responding to a scammer. "
                    f"The scam type is {scam}. "
                    f"Respond in a {style} manner. "
                    f"Do NOT give any sensitive information. "
                    f"Ask questions, delay, or gently confuse the scammer. "
                    f"Sound completely human and natural."
                )
            }
        ]
    })

# -------- NORMAL MYRA DATA --------
for _ in range(normal_samples):
    prompt = random.choice(normal_prompts)
    topic = random.choice(normal_topics)
    tone = random.choice(normal_tones)

    dataset.append({
        "messages": [
            {
                "role": "system",
                "content": MYRA_SYSTEM
            },
            {
                "role": "user",
                "content": f"{prompt} The topic is {topic}."
            },
            {
                "role": "assistant",
                "content": (
                    f"Respond as Myra in a {tone} and natural way. "
                    f"Be empathetic, human, and realistic."
                )
            }
        ]
    })

# ---------------- SAVE FINAL DATASET ----------------
with open("myra_final_spam_conversation_dataset_4000.json", "w") as f:
    json.dump(dataset, f, indent=2)

print("‚úÖ FINAL DATASET CREATED")
print(f"üìä Total samples: {len(dataset)}")
print(f"üõ°Ô∏è Spam-defense samples: {spam_samples}")
print(f"üí¨ Normal conversation samples: {normal_samples}")
print("üìÅ File: myra_final_spam_conversation_dataset_4000.json")
````

````python
from datasets import load_dataset

raw_dataset = load_dataset(
    "json",
    data_files="myra_final_spam_conversation_dataset_4000.json",
    split="train"
)

def flatten_conversation(example):
    text = ""
    for msg in example["messages"]:
        if msg["role"] == "system":
            text += f"<|system|>\n{msg['content']}\n"
        elif msg["role"] == "user":
            text += f"<|user|>\n{msg['content']}\n"
        elif msg["role"] == "assistant":
            text += f"<|assistant|>\n{msg['content']}\n"
    return {"text": text.strip()}

dataset = raw_dataset.map(
    flatten_conversation,
    remove_columns=raw_dataset.column_names,
)

print(dataset)
print(dataset[0]["text"][:500])
````

````python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

spam_data = [
    ("Your bank account will be blocked", 1),
    ("Send OTP immediately", 1),
    ("Urgent verification required", 1),
    ("Crypto investment guaranteed returns", 1),
    ("Job offer earn money fast", 1),

    ("Hey how was your day?", 0),
    ("Are we meeting today?", 0),
    ("I feel tired today", 0),
    ("Can you help me with studies?", 0),
]

texts = [x[0] for x in spam_data]
labels = [x[1] for x in spam_data]

vectorizer = TfidfVectorizer(ngram_range=(1,2), stop_words="english")
X = vectorizer.fit_transform(texts)

spam_model = LogisticRegression(class_weight="balanced", max_iter=300)
spam_model.fit(X, labels)

def is_spam(msg: str) -> bool:
    return spam_model.predict(vectorizer.transform([msg]))[0] == 1
````

````python
conversation = {
    "is_spam": None,
    "history": []
}

def reset_conversation():
    conversation["is_spam"] = None
    conversation["history"] = []

def add_turn(role, text):
    conversation["history"].append(f"{role}: {text}")

def get_context(max_turns=4):
    return "\n".join(conversation["history"][-max_turns:])
````

````python
def scammer_messages(message):
    return [
        {
            "role": "system",
            "content": (
                "You are Myra, a real human woman. "
                "You are talking to a scammer. "
                "Do not give sensitive information. "
                "Be polite, calm, slightly confused, and delay naturally. "
                "Sound completely human."
            )
        },
        {"role": "user", "content": message}
    ]

def normal_messages(message):
    return [
        {
            "role": "system",
            "content": (
                "You are Myra, a friendly 23-year-old woman chatting casually. "
                "Speak like a real person talking to a friend. "
                "Do not offer help unless asked. "
                "Do not sound like customer support. "
                "Do not mention being an AI."
            )
        },
        {"role": "user", "content": message}
    ]

````

````python
def myra_agent(message: str):
    if conversation["is_spam"] is None:
        conversation["is_spam"] = is_spam(message)

    if conversation["is_spam"]:
        messages = scammer_messages(message)
        role = "Scammer"
    else:
        messages = normal_messages(message)
        role = "User"

    reply = myra_generate(messages)

    add_turn(role, message)
    add_turn("Myra", reply)

    return reply
````

### Data Prep
We now use the `Llama-3.1` format for conversation style finetunes. We use [Maxime Labonne's FineTome-100k](https://huggingface.co/datasets/mlabonne/FineTome-100k) dataset in ShareGPT style. But we convert it to HuggingFace's normal multiturn format `("role", "content")` instead of `("from", "value")`/ Llama-3 renders multi turn conversations like below:

```
<|begin_of_text|><|start_header_id|>user<|end_header_id|>

Hello!<|eot_id|><|start_header_id|>assistant<|end_header_id|>

Hey there! How are you?<|eot_id|><|start_header_id|>user<|end_header_id|>

I'm great thanks!<|eot_id|>
```

We use our `get_chat_template` function to get the correct chat template. We support `zephyr, chatml, mistral, llama, alpaca, vicuna, vicuna_old, phi3, llama3` and more.

````python
from unsloth.chat_templates import get_chat_template

tokenizer = get_chat_template(
    tokenizer,
    chat_template = "llama-3.1",
)

def formatting_prompts_func(examples):
    convos = examples["conversations"]
    texts = [tokenizer.apply_chat_template(convo, tokenize = False, add_generation_prompt = False) for convo in convos]
    return { "text" : texts, }

from datasets import load_dataset
dataset = load_dataset("mlabonne/FineTome-100k", split = "train")
````

We now use `standardize_sharegpt` to convert ShareGPT style datasets into HuggingFace's generic format. This changes the dataset from looking like:
```
{"from": "system", "value": "You are an assistant"}
{"from": "human", "value": "What is 2+2?"}
{"from": "gpt", "value": "It's 4."}
```
to
```
{"role": "system", "content": "You are an assistant"}
{"role": "user", "content": "What is 2+2?"}
{"role": "assistant", "content": "It's 4."}
```

````python
from unsloth.chat_templates import standardize_sharegpt
dataset = standardize_sharegpt(dataset)
dataset = dataset.map(formatting_prompts_func, batched = True,)
````

### Train the model
Now let's train our model. We do 60 steps to speed things up, but you can set `num_train_epochs=1` for a full run, and turn off `max_steps=None`. We also support TRL's `DPOTrainer`!

````python
from trl import SFTConfig, SFTTrainer
from transformers import DataCollatorForSeq2Seq
trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = dataset,
    dataset_text_field = "text",
    max_seq_length = max_seq_length,
    data_collator = DataCollatorForSeq2Seq(tokenizer = tokenizer),
    packing = False, # Can make training 5x faster for short sequences.
    args = SFTConfig(
        per_device_train_batch_size = 2,
        gradient_accumulation_steps = 4,
        warmup_steps = 5,
        # num_train_epochs = 1, # Set this for 1 full training run.
        max_steps = 60,
        learning_rate = 2e-4,
        logging_steps = 1,
        optim = "adamw_8bit",
        weight_decay = 0.001,
        lr_scheduler_type = "linear",
        seed = 3407,
        output_dir = "outputs",
        report_to = "none", # Use TrackIO/WandB etc
    ),
)
````

We also use Unsloth's `train_on_completions` method to only train on the assistant outputs and ignore the loss on the user's inputs.

````python
from unsloth.chat_templates import train_on_responses_only
trainer = train_on_responses_only(
    trainer,
    instruction_part = "<|start_header_id|>user<|end_header_id|>\n\n",
    response_part = "<|start_header_id|>assistant<|end_header_id|>\n\n",
)
````

### GGUF / llama.cpp Conversion
To save to `GGUF` / `llama.cpp`, we support it natively now! We clone `llama.cpp` and we default save it to `q8_0`. We allow all methods like `q4_k_m`. Use `save_pretrained_gguf` for local saving and `push_to_hub_gguf` for uploading to HF.

Some supported quant methods (full list on our [Wiki page](https://github.com/unslothai/unsloth/wiki#gguf-quantization-options)):
* `q8_0` - Fast conversion. High resource use, but generally acceptable.
* `q4_k_m` - Recommended. Uses Q6_K for half of the attention.wv and feed_forward.w2 tensors, else Q4_K.
* `q5_k_m` - Recommended. Uses Q6_K for half of the attention.wv and feed_forward.w2 tensors, else Q5_K.

[**NEW**] To finetune and auto export to Ollama, try our [Ollama notebook](https://colab.research.google.com/github/unslothai/notebooks/blob/main/nb/Llama3_(8B)-Ollama.ipynb)

````python
# Save to 8bit Q8_0
if False: model.save_pretrained_gguf("model", tokenizer,)
# Remember to go to https://huggingface.co/settings/tokens for a token!
# And change hf to your username!
if False: model.push_to_hub_gguf("hf/model", tokenizer, token = "")

# Save to 16bit GGUF
if False: model.save_pretrained_gguf("model", tokenizer, quantization_method = "f16")
if False: model.push_to_hub_gguf("hf/model", tokenizer, quantization_method = "f16", token = "")

# Save to q4_k_m GGUF
if False: model.save_pretrained_gguf("model", tokenizer, quantization_method = "q4_k_m")
if False: model.push_to_hub_gguf("hf/model", tokenizer, quantization_method = "q4_k_m", token = "")

# Save to multiple GGUF options - much faster if you want multiple!
if False:
    model.push_to_hub_gguf(
        "hf/model", # Change hf to your username!
        tokenizer,
        quantization_method = ["q4_k_m", "q8_0", "q5_k_m",],
        token = "", # Get a token at https://huggingface.co/settings/tokens
    )
````
